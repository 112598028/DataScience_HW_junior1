```python
# 匯入Matplotlib與Seaborn
# Seaborn能讓圖表更美麗
import matplotlib as mpl
import seaborn as sns

# 讓pyplot能以plt的別名來執行
import matplotlib.pyplot as plt

# Jupyter Notebook上顯示圖表所必須的Magic Command
%matplotlib inline

import numpy as np
import numpy.random as random
```

```python
def my_function(x):
    return 5*x + 3

x = np.arange(-10, 10)
plt.figure(figsize = (20, 6))
plt.plot(x, my_function(x))
plt.grid(True)
```

```python
x = np.linspace(-10, 10, 100)

plt.figure(figsize = (20, 6))
plt.plot(x, np.sin(x), label='sin')
plt.plot(x, np.cos(x), label='cos')
plt.legend()
plt.grid(True)
```

```python
plt.figure(figsize = (10, 8))

random.seed(0)

plt.subplot(1, 2, 1) 
plt.hist(np.random.random(1000))

plt.subplot(1, 2, 2)
plt.hist(np.random.random(1000))

plt.show()
```

```python
import pandas as pd

student_data_math = pd.read_csv('student-mat.csv', sep=";")

student_data_math.head()
```

```python
student_data_math['reason'].head(15)
```

```python
from pandas import Series, DataFrame

df_math = DataFrame(student_data_math)

df_math_reason = df_math['reason']
```

```python
data = df_math_reason.value_counts()

data
```

```python
labels = ['course', 'reputation', 'home', 'other']

colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']

plt.figure(figsize = (15, 6))

plt.pie(data , labels = labels, colors = colors, autopct = '%1.1f%%', shadow = True, startangle = 90)

plt.axis('equal')

plt.title('Student Math reason')

plt.show()
```

```python
student_data_math['higher'].head(15)
```

```python
df_math_higher_sc = df_math.groupby('higher')['G3'].mean()

df_math_higher_sc
```

```python
x = [1, 2]
plt.figure(figsize = (10, 6))

plt.bar(x, df_math_higher_sc, align = 'center')

plt.xticks(x, ['no','yes'])
plt.ylabel('G3 Score mean')
plt.xlabel('higher')

plt.grid(True)
```

```python
student_data_math['traveltime'].head(15)
```

```python
df_math_traveltime_sc = df_math.groupby('traveltime')['G3'].mean()

df_math_traveltime_sc
```

```python
x = [1, 2, 3, 4]
plt.figure(figsize = (10, 6))

plt.barh(x, df_math_traveltime_sc, align = 'center')

plt.yticks(x, ['1','2', '3', '4'])
plt.ylabel('traveltime')
plt.xlabel('G3 Score mean')

plt.grid(True)
```
